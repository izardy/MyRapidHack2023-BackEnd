```python
infoDepot=pd.read_csv('../data/processed/InfoDEPOT.csv',low_memory=False)[['depot_id','depot_name']].drop_duplicates()

infoTRIP=pd.read_csv('../data/trip_departure/processed/td_agg.csv',low_memory=False)
infoTRIP['ope_date']=pd.to_datetime(infoTRIP['ope_date']+ ' 00:00:00').dt.strftime('%Y-%m-%d %H:%M:%S')

infoBas=pd.read_csv('../data/processed/InfoBAS.csv')

infoBas['ope_date']=pd.to_datetime(infoBas['ope_date']+ ' 00:00:00').dt.strftime('%Y-%m-%d %H:%M:%S')
infoBas['br_date']=pd.to_datetime(infoBas['br_date']+ ' 00:00:00').dt.strftime('%Y-%m-%d %H:%M:%S')
infoBas['br_date_entry']=pd.to_datetime(infoBas['br_date_entry']+ ' 00:00:00').dt.strftime('%Y-%m-%d %H:%M:%S')
infoBas['br_target_completed']=pd.to_datetime(infoBas['br_target_completed']+ ' 00:00:00').dt.strftime('%Y-%m-%d %H:%M:%S')
infoBas['rtd_date']=pd.to_datetime(infoBas['rtd_date']+ ' 00:00:00').dt.strftime('%Y-%m-%d %H:%M:%S')
infoBas['date_issues']=pd.to_datetime(infoBas['date_issues']+ ' 00:00:00').dt.strftime('%Y-%m-%d %H:%M:%S')
infoBas['acc_date']=pd.to_datetime(infoBas['acc_date']+ ' 00:00:00').dt.strftime('%Y-%m-%d %H:%M:%S')

#infoBas=infoBas.drop(columns=['label_1','label_2','label_3'])

infoBas=pd.merge(infoBas,infoTRIP,how='left',on=['ope_date','depot_id','line_id','bus_no']).drop_duplicates()
infoBas=pd.merge(infoBas,infoDepot,how='left',on=['depot_id']).drop_duplicates()


#imputation
infoBas['br_id']=infoBas['br_id'].fillna('0000000').astype(str).str.replace('\.0','')
infoBas['br_repair_status']=infoBas['br_repair_status'].fillna('N/A').astype(str)
infoBas['br_entry_status']=infoBas['br_entry_status'].fillna('N/A').astype(str)
infoBas['br_date']=infoBas['br_date'].fillna('N/A')
infoBas['br_date_entry']=infoBas['br_date_entry'].fillna('N/A')
infoBas['br_target_completed']=infoBas['br_target_completed'].fillna('N/A')
infoBas['br_ttl_days']=pd.to_numeric(infoBas['br_ttl_days'],errors='coerce').fillna(0).astype(int)
#infoBas['br_repair_status']=pd.to_numeric(infoBas['br_repair_status'], errors='coerce').fillna('0000000').astype(str).str.replace('\.0','')
infoBas['rtd_date']=infoBas['rtd_date'].fillna('N/A')
infoBas['rtd_id']=pd.to_numeric(infoBas['rtd_id'], errors='coerce').fillna(0).astype(int)
infoBas['route_id']=pd.to_numeric(infoBas['route_id'], errors='coerce').fillna(0).astype(int)
infoBas['bus_model_id']=pd.to_numeric(infoBas['bus_model_id'], errors='coerce').fillna(0).astype(int)
infoBas['rtd_cause']=pd.to_numeric(infoBas['rtd_cause'], errors='coerce').fillna(0).astype(int)

infoBas['driver_id']=infoBas['driver_id'].fillna('N/A')
infoBas['cause_remark']=infoBas['cause_remark'].fillna('N/A')
infoBas['fc_remark']=infoBas['fc_remark'].fillna('N/A')

infoBas['date_issues']=infoBas['date_issues'].fillna('N/A')
infoBas['capt_id']=infoBas['capt_id'].fillna('N/A')
infoBas['bcc_acc_id']=infoBas['bcc_acc_id'].fillna('N/A')
infoBas['acc_date']=infoBas['acc_date'].fillna('N/A')
infoBas['bus_damage']=infoBas['bus_damage'].fillna('N/A')

infoBas['eng_find']=infoBas['eng_find'].fillna('N/A')

infoBas['label_1']=infoBas['label_1'].fillna('N/A').str.upper()
infoBas['label_2']=infoBas['label_2'].fillna('N/A').str.upper()
infoBas['label_3']=infoBas['label_3'].fillna('N/A').str.upper()
infoBas['label']=infoBas['label'].fillna('N/A').str.upper()

infoBas['point_a']=infoBas['point_a'].str.upper()
infoBas['point_b']=infoBas['point_b'].str.upper()

infoBas['operational_issue']=np.where(infoBas['label']=='NORMAL TRIP','NO','YES')

infoBas['days_before_issue']=np.where(infoBas['operational_issue']=='YES',0,np.nan)

#infoBas.to_csv('../data/model/data_model.csv',index=False)

infoBas#.columns

#/home/hadoop/MyRapidHack2023-BackEnd/

```
